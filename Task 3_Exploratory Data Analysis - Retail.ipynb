## Data Visualization

```python
plt.bar(df['Quantity'], df['Profit'])
plt.title(' Profit Vs Quantity', fontsize=14)
plt.xlabel('Quantity', fontsize=14)
plt.ylabel('Profit', fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.bar(df['Quantity'], df['Sales'], color="red")
plt.title(' Quantity Vs Sales', fontsize=14)
plt.xlabel('Quantity', fontsize=14)
plt.ylabel('Sales', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(16,8))

sns.relplot( x ='Quantity',y='Profit',color='purple',hue="Segment" , data=df )
plt.show()
```

* Most of the products were bought in quantity of 5.

```python
sns.countplot(x = 'Segment',data = df, palette="coolwarm")
plt.show()
```

```python
plt.title("Segment", fontsize=20)
df['Segment'].value_counts().plot.pie(autopct ='% 1.1f %%', shadow = True)
plt.show()
```

* Majority of products are boughts by consumer segment.


```python
sns.countplot(x='Category',data=df,palette='tab10')
plt.show()
```

* Most bought products are frm office supllies category.

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['green','orange'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

* Consumer segmenthave have high Sales.
* All three segments have approximately same profit

```python
plt.title("Region", fontsize=20)
df['Region'].value_counts().plot.pie(autopct ='% 1.1f %%', shadow = True)
plt.show()
```

* West region has done most of sales.

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='coolwarm_r')
plt.xlabel('Region',fontsize=15)
plt.show()
```

* Segment wise sales are almost same in every region.

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['blue','skyblue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.countplot(x = "State", data = df, palette='rocket_r',order=df['State'].value_counts().index);
plt.xlabel('States')
plt.xticks(rotation = 70)
plt.show()
```

* Highest number of buyers are from Calofornia and New York.
* Lowest number of buyers are from west Virginia and Wyoming.

```python
n_data=['Sales','Quantity','Discount','Profit']
plt.figure(figsize=(10,5))
sns.heatmap(df[n_data].corr(),annot=True, fmt='.4f',cmap='coolwarm_r',center=0 );
```

* There is negative corelation betweem Profit and  Discount.
* Wherer as ok corelation between sales and profit.

```python
sns.pairplot(df,hue='Category')
plt.show()
```

* East and West region have high profit and sales.
